```python
import cv2
import numpy as np
import os
from skimage import restoration


def enhance_image(img):

    # 1. White Balance + CLAHE
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)

    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
    cl = clahe.apply(l)

    limg = cv2.merge((cl,a,b))
    result = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)

    # 2. Denoise
    result = cv2.fastNlMeansDenoisingColored(result,None,10,10,7,21)

    # 3. Sharpen
    kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
    result = cv2.filter2D(result,-1,kernel)

    return result


def enhance_folder(input_dir, output_dir):

    os.makedirs(output_dir, exist_ok=True)

    for img_name in os.listdir(input_dir):

        if not img_name.lower().endswith((".jpg",".png",".jpeg")):
            continue

        path = os.path.join(input_dir, img_name)

        img = cv2.imread(path)

        if img is None:
            continue

        enhanced = enhance_image(img)

        cv2.imwrite(os.path.join(output_dir, img_name), enhanced)


# ✅ Process dataset (NO ../ now)
enhance_folder("under-water-1/train/images", "enhanced/train")
enhance_folder("under-water-1/valid/images", "enhanced/valid")
enhance_folder("under-water-1/test/images", "enhanced/test")

print("Enhancement Completed ✅")
```

```python
import cv2
import numpy as np
import os
from tqdm import tqdm


def enhance_image_fast(img):

    # CLAHE
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)

    clahe = cv2.createCLAHE(2.0, (8,8))
    cl = clahe.apply(l)

    result = cv2.merge((cl,a,b))
    result = cv2.cvtColor(result, cv2.COLOR_LAB2BGR)

    # Sharpen (fast)
    kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
    result = cv2.filter2D(result,-1,kernel)

    return result


def enhance_folder_fast(input_dir, output_dir):

    os.makedirs(output_dir, exist_ok=True)

    images = [f for f in os.listdir(input_dir)
              if f.lower().endswith((".jpg",".png",".jpeg"))]

    for img_name in tqdm(images):

        img = cv2.imread(os.path.join(input_dir,img_name))

        if img is None:
            continue

        enhanced = enhance_image_fast(img)

        cv2.imwrite(os.path.join(output_dir,img_name), enhanced)


enhance_folder_fast("under-water-1/train/images", "enhanced/train")
enhance_folder_fast("under-water-1/valid/images", "enhanced/valid")
enhance_folder_fast("under-water-1/test/images", "enhanced/test")

print("Fast Enhancement Done ✅")
```

```python
import cv2
import numpy as np
import os
from tqdm import tqdm


def enhance_image_fast(img):

    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)

    clahe = cv2.createCLAHE(2.0, (8,8))
    cl = clahe.apply(l)

    result = cv2.merge((cl,a,b))
    result = cv2.cvtColor(result, cv2.COLOR_LAB2BGR)

    kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
    result = cv2.filter2D(result,-1,kernel)

    return result


def enhance_train_only():

    input_dir = "under-water-1/train/images"
    output_dir = "enhanced/train"

    os.makedirs(output_dir, exist_ok=True)

    images = [f for f in os.listdir(input_dir)
              if f.lower().endswith((".jpg",".png",".jpeg"))]

    for img_name in tqdm(images, desc="Enhancing TRAIN"):

        img = cv2.imread(os.path.join(input_dir, img_name))

        if img is None:
            continue

        enhanced = enhance_image_fast(img)

        cv2.imwrite(os.path.join(output_dir, img_name), enhanced)


enhance_train_only()

print("Train Enhancement Done ✅")
```

```python
import cv2
import os

input_dir = "waterweeds_only"
output_dir = "waterweeds_aug"

os.makedirs(output_dir, exist_ok=True)

count = 0

for img_name in os.listdir(input_dir):
    img_path = os.path.join(input_dir, img_name)
    img = cv2.imread(img_path)

    if img is None:
        continue

    # Flip
    flip = cv2.flip(img, 1)

    # Rotate
    rot = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

    # Brightness
    bright = cv2.convertScaleAbs(img, alpha=1.2, beta=30)

    # Save
    cv2.imwrite(f"{output_dir}/flip_{count}.jpg", flip)
    cv2.imwrite(f"{output_dir}/rot_{count}.jpg", rot)
    cv2.imwrite(f"{output_dir}/bright_{count}.jpg", bright)

    count += 1

print("Augmentation done ✅", count)
```

```python
import os
import shutil

aug_dir = "waterweeds_aug"
label_dir = "under-water-1/train/labels"

copied = 0
skipped = 0

for img in os.listdir(aug_dir):

    # Example: flip_12.jpg → 12.jpg
    parts = img.split("_")

    if len(parts) < 2:
        skipped += 1
        continue

    orig = parts[1]   # original filename number
    src_label = os.path.join(label_dir, orig.replace(".jpg", ".txt"))
    new_label = os.path.join(label_dir, img.replace(".jpg", ".txt"))

    if os.path.exists(src_label):
        shutil.copy(src_label, new_label)
        copied += 1
    else:
        skipped += 1

print("Copied labels:", copied)
print("Skipped:", skipped)
```

```python
import cv2
import os

input_dir = "waterweeds_only"
output_dir = "waterweeds_aug"

os.makedirs(output_dir, exist_ok=True)

count = 0

for img_name in os.listdir(input_dir):

    img_path = os.path.join(input_dir, img_name)
    img = cv2.imread(img_path)

    if img is None:
        continue

    name = img_name.replace(".jpg","")

    # Flip
    flip = cv2.flip(img, 1)

    # Rotate
    rot = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

    # Brightness
    bright = cv2.convertScaleAbs(img, alpha=1.2, beta=30)

    cv2.imwrite(f"{output_dir}/{name}_flip.jpg", flip)
    cv2.imwrite(f"{output_dir}/{name}_rot.jpg", rot)
    cv2.imwrite(f"{output_dir}/{name}_bright.jpg", bright)

    count += 3

print("Correct augmentation done ✅ Images:", count)
```

```python
import os
import shutil

img_dir = "under-water-1/train/images"
label_dir = "under-water-1/train/labels"

fixed = 0
skipped = 0

for img in os.listdir(img_dir):

    if not img.endswith(".jpg"):
        continue

    txt = img.replace(".jpg", ".txt")
    dst = os.path.join(label_dir, txt)

    # If label missing
    if not os.path.exists(dst):

        # Remove augmentation prefix
        base = img

        for prefix in ["bright_", "flip_", "rot_"]:
            if base.startswith(prefix):
                base = base.replace(prefix, "", 1)

        src = os.path.join(label_dir, base.replace(".jpg", ".txt"))

        # If original label exists → copy it
        if os.path.exists(src):
            shutil.copy(src, dst)
            fixed += 1
        else:
            skipped += 1

print("Fixed labels:", fixed)
print("Still missing:", skipped)
```

```python
import os

img_dir = "under-water-1/train/images"
label_dir = "under-water-1/train/labels"

removed = 0

for img in os.listdir(img_dir):

    if not img.endswith(".jpg"):
        continue

    txt = img.replace(".jpg", ".txt")

    if not os.path.exists(os.path.join(label_dir, txt)):
        os.remove(os.path.join(img_dir, img))
        removed += 1

print("Removed images without labels:", removed)
```

```python
import pandas as pd

rows = []

for r in results:
    img_name = r.path.split("/")[-1]

    if r.boxes is not None:
        for box in r.boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            x1, y1, x2, y2 = box.xyxy[0].tolist()

            rows.append([
                img_name,
                cls,
                conf,
                x1, y1, x2, y2
            ])

# Create DataFrame
df = pd.DataFrame(rows, columns=[
    "image", "class_id", "confidence", "x1", "y1", "x2", "y2"
])

# Save CSV
df.to_csv("test_results.csv", index=False)

print("CSV File Saved: test_results.csv ✅")
```
